### Part (a) [\*]

Define a function called `print_alternate_lines` that takes in the name of a file as its parameter. The function prints out alternate lines from the file, starting from the first line. That is, the function prints out the first line, the third line, the fifth line, etc. of the file until the file ends.

For example, if the file looks like the following:

```
1
22
333
4444
4444
333
22
1
```

The function should print out the following output:

```
1
333
4444
22
```

````python
# Write your code here:
def print_alternate_lines(file): 
    lines_list = []
    with open(file, 'r') as my_file:
        for line in my_file:
            line = line.rstrip('\n')
            lines_list.append(line)
    for i in range(len(lines_list)): 
        if i % 2 == 0: 
            print (lines_list[i])
````

````python
# Write your code here:
def get_universities_founded_before(file, year): 
    output = []
    with open(file, 'r') as my_file: 
        for line in my_file: 
            line = line.rstrip("\n")
            columns = line.split("\t")
            name_of_a_university = columns[0]
            year_founded = int(columns[2])
            if year_founded < year: 
                output.append(name_of_a_university)
    return output

# for return at the end of a function with file opening, does it matter if I add another indent? 
````

Define a function called ```get_average_num_students()```. This function takes in the name of a file as described above and returns the average number of undergraduate students of all the universities in the given file. If the file is empty, the function returns 0.0.

````python
# Write your code here:
def get_average_num_students(file): 
    total = 0
    count = 0
    with open(file, 'r') as my_file:
        for line in my_file: 
            line = line.rstrip("\n")
            columns = line.split("\t")
            number_of_undergraduate_students = int(columns[3])
            count += 1 
            total += number_of_undergraduate_students
        average = total / count  
    return average
````

Define a function called ```search_with_keyword()```. This function takes in the name of a file as described above and a string that represents a keyword. The function returns a list of strings that represent the __acronyms__ of the universities whose __full names__ contain the given keyword.

For example, given the file as shown above, if the keyword is ```'California'```, then the function should return ```['Berkeley', 'UCLA']```.

Your search needs to be __case-insensitive__. So searching for ```'california'``` or ```'California'``` should give you the same results.

````python
# Write your code here:
def search_with_keyword(file, keyword): 
    keyword = keyword.lower()
    output = []
    with open(file, 'r') as my_file:
        for line in my_file: 
            line = line.rstrip("\n")
            columns = line.split("\t")
            name_of_a_university = columns[0].lower()
            acronym = columns[1]
            if keyword in name_of_a_university: 
                output.append(acronym)
    return output
````

## Q3: Temperatures [\*\*]

You are given a file called `'temperatures.txt'` that stores some patients' temperature readings. Each line of the file is about a single patient. The patient's name is displayed first followed by a sequence of temperature values. The name and the different temperature values are separated by tabs. Note that the numbers of temperature readings are different for different patients.

Read in this file and print out the minimum and maximum temperature readings of each patient.

__You can assume that for each patient, there is at least one temperature reading in the file.__

Your program should display the following output:

```
Kate: 36.3, 37.6
Larry: 36.1, 38.4
Jerry: 37.6, 38.9
```

````python
# Write your code here:
with open('temperatures.txt', 'r') as my_file:
    for line in my_file:
        minimum = 100.0
        maximum = 0.0
        line = line.rstrip('\n')
        columns = line.split('\t')
        name = columns[0]
        for i in range(1, len(columns)):
            temperature = float(columns[i])
            if temperature <= minimum: 
                minimum = temperature
            if temperature >= maximum:
                maximum = temperature
        print (name + ": " + str(minimum) + ", " + str(maximum))
````

## Q5: Matrix

In mathematics, a square matrix refers to a square array of numbers. For example, the following is a 3 $\times$ 3 square matrix:

$$
\begin{bmatrix}
3 & 9 & 5 \\
2 & 4 & 1 \\
8 & 2 & 7 \\
\end{bmatrix}
$$

### Part (a) [\*\*]

Define a function called ```get_matrix``` that takes in a string, which is the name of a file that stores such a square matrix. For example, to store an $n \times n$ square matrix, the file has $n$ rows, where each row contains $n$ numbers separated by spaces.

The function returns a __list of lists of numbers__, representing the square matrix stored in the file.

````python
# Write your code here:
def get_matrix(file): 
    matrix_list = []
    with open(file, 'r') as my_file: 
        for line in my_file:
            line = line.rstrip('\n')
            # this gives a list of integers
            row_numbers_list = line.split(' ')
            # this gives a list of floats
            row_floats_list = []
            for number in row_numbers_list: 
                number = float(number)
                row_floats_list.append(number)
            matrix_list.append(row_floats_list)
    return matrix_list

# is there any particular reason why the numbers are converted from integers to floats? 
````

## Q6: [\*\*\*]

In this question, you need to read in a Python script and extract all the docstrings together with their function headers.

Assume the Python script you're processing contains several function defintions. For each function defintion, assume it is strictly written in the following way:

- The function header (i.e., ```def name_of_function(list_of_parameters):```) takes one line.
- Immediately below the function header, a docstring is presented to explain the function. 
    - The docstring may take one line or multiple lines. 
    - It always starts with ```"""``` (i.e., three double quotes), but there are 4 spaces in front of the ```"""``` for indentation purpose.
    - It always ends with ```"""```, but there may be some additional spaces behind the ```"""```.

For example, a file may look like the following:

```
def my_test_func(param1, param2):
    """
    This is a test function that takes in two parameters and returns a number.
    The two parameters are compared. 0 or 1 is returned.
    """
    if (param1 < param2):
        return 0
    else:
        return 1
def another_function():
    """This is a function without any parameter."""
    pass
def third_funct():
    """This is a function
    that has two lines of docstring."""
    print("hello")
```

You can assume that inside the Python script you're processing, the keyword ```def``` occurs only for function definitions. Also, the pattern ```"""``` does not occur anywhere else in the file except for docstrings.

Write a function called ```extract_functions()``` that takes in two parameters:
- ```python_filename```: The name of a Python script, with .py as its extension.
- ```output_filename```: The name of an output file.
    
The function should read the Python script, extracts all function headers and docstrings, and write them into the output file. In the output file, the functions are numbered. The doctrings are indented with 4 spaces. The keyword ```def```, the colon, and ```"""``` are removed. A blank line is inserted after each function.

For example, given the file above, the generated output file should contain the following:

```
1. my_test_func(param1, param2)
    This is a test function that takes in two parameters and returns a number.
    The two parameters are compared. 0 or 1 is returned.

2. another_function()
    This is a function without any parameter.

3. third_funct()
    This is a function
    that has two lines of docstring.
```

You are given a file called "python_script.py" that can be used to test your program.

````python
# Write your code here:
def extract_functions(python_filename, output_filename):
    # create a list to store all lines in file 
    lines_list = []
    triplequote = '"""'
    with open(python_filename, 'r') as input_file:  
        for line in input_file:
            line = line.rstrip('\n')
            lines_list.append(line)
    
    # create a list to store the indices of lines which contain a function
    function_indices = []
    # create a list to store the indices of lines which contain """
    docstring_indices = []
    for x in range(len(lines_list)):
        line = lines_list[x]
        if 'def ' in line: 
            function_indices.append(x)
        elif triplequote in line:
            if line.count(triplequote) == 1: 
                docstring_indices.append(x)
            # this is to deal with the scenario where docstring is one line; the start and end index are the same
            # if not stored in pairs, there will be problems accessing the indices later
            elif line.count(triplequote) == 2:
                docstring_indices.append(x)
                docstring_indices.append(x)
    
    # create a list of functions
    functions = []
    for function_index in function_indices:
        function = lines_list[function_index].strip('def ')
        functions.append(function)
    # create a list of docstrings
    docstrings = []
    
    # retrieve the start and end indices of lines containing docstrings 
    for y in range(0, len(docstring_indices), 2): 
        start = docstring_indices[y]
        end = docstring_indices[y+1]
        docstring = []
        for z in range(len(lines_list)):
            # add all lines between the start and end indices of lines with docstrings sans docstrings
            if start <= z <= end:
                # get rid of triplequote at the start and end of docstring
                line = lines_list[z].strip(triplequote)
                line = line.strip('    ' + triplequote)
                docstring.append(line)
        docstrings.append(docstring)
    
    with open(output_filename, 'w') as output_file:
        for a in range(len(functions)):
            function_number = str(a + 1)
            function = functions[a]
            docstring = docstrings[a]
            output_file.write( function_number + ". " + function + '\n' )
            for b in range(len(docstring)):
                docstring_line = docstring[b]
                # check whether the string is empty; if yes no need to print
                if docstring_line != '': 
                    output_file.write('\t' + docstring_line + '\n' )
            # insert a blank line to space out functions
            output_file.write( " " + '\n' )
````
