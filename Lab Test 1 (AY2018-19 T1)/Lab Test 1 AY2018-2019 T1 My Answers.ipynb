```python
# Question 2
# Name: Rozalin Mohanty
# Email ID: rozalin.mohanty.2017@law.smu.edu.sg 
#
def represent_numbers (start, end):
    # write your answer between #startcode and #endcode
    #startcode
    output = ""
    if end < start: 
        output = ""
    else:
        for n in range(start,end):
            m = abs(n)
            output += (m*"-" + "#")
        output += (abs(end) * "-")
    return output
    #endcode

print('Test 1')
print('Expected:-#--#---#----#-----')
print('Actual  :' + represent_numbers(1, 5))
print()

print('Test 2')
print('Expected:---#----#-----')
print('Actual  :' + represent_numbers(3, 5))
print()

print('Test 3')
print('Expected:-')
print('Actual  :' + represent_numbers(1, 1))
print()

print('Test 4')
print('Expected:[]')
print('Actual  :[' + represent_numbers(4, 1) + ']')
print()

print('Test 5')
print('Expected:----#---#--#-##-')
print('Actual  :' + represent_numbers(-4, 1))
print()
```

```python
# Question 3
# Name: Rozalin Mohanty
# Email ID: rozalin.mohanty.2017@law.smu.edu.sg 
#
def mask_out(sentence, banned, substitutes):
    # write your answer between #start and #end
    #start
    output = ""
    if len(banned) == len(substitutes): 
        for ch in sentence: 
            for i in range(len(banned)): 
                if ch == banned[i]:
                    output += substitutes[i]
                else:
                    if ch not in output and ch not in banned:
                        output += ch 
    else: 
        for ch in sentence: 
            for i in range(len(banned)): 
                if ch == banned[i]:
                    if i < len(substitutes): 
                            output += substitutes[i]
                    else: 
                        output += substitutes[0]
                else:
                    if ch not in output and ch not in banned:
                        output += ch
    return output
    #end

# you only want the ch to be printed if it wasn't already printed and it's not a banned letter
# if it was a banned letter, it wouldn't have ever been printed but you don't want it to be printed
    
print('Test 1')
print('Expected:abcd#')
print('Actual  :' + mask_out('abcde', 'e', '#'))
print()

print('Test 2')
print('Expected:#$solute')
print('Actual  :' + mask_out('absolute', 'ab', '#$'))
print()

print('Test 3')
print('Expected:121hon')
print('Actual  :' + mask_out('python', 'pyt', '12'))
print()
```
