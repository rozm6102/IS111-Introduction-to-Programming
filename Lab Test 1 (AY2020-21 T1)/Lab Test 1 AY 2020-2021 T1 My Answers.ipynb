```python
def get_sum_of_digits(my_str):
    # Replace the code below with your implementation.
    output = 0
    integers = "0123456789"
    # for the case of an empty string
    if my_str == "":
        output = 0
    # for the case where the string actually contains something 
    # test whether it's an integer before adding 
    else: 
        for ch in my_str: 
            if ch in integers: 
                output += int(ch)
        
    return output 

print()
print('-' * 20)
print()

print('Test Case 1: get_sum_of_digits("1234")')

print()

result = get_sum_of_digits("1234")
print('Expected: 10')
print('Actual:   ' + str(result))

print("Expected type of returned value: <class 'int'>")
print('Actual type of returned value:   ' + str(type(result)))

print()
print('-' * 20)
print()

print('Test Case 2: get_sum_of_digits("SMU1-2-3-4")')

print()

result = get_sum_of_digits("SMU1-2-3-4")
print('Expected: 10')
print('Actual:   ' + str(result))

print()
print('-' * 20)
print()

print('Test Case 3: get_sum_of_digits("32-4*5(0)36")')

print()

result = get_sum_of_digits("32-4*5(0)36")
print('Expected: 23')
print('Actual:   ' + str(result))

print()
print('-' * 20)
print()

print('Test Case 4: get_sum_of_digits("SMU-SIS")')

print()

result = get_sum_of_digits("SMU-SIS")
print('Expected: 0')
print('Actual:   ' + str(result))

print()
print('-' * 20)
print()

print('Test Case 5: get_sum_of_digits("")')

print()

result = get_sum_of_digits("")
print('Expected: 0')
print('Actual:   ' + str(result))

print()
print('-' * 20)
print()
```

```python
def check_math(list_of_equations):
    # Replace the code below with your implementation.
    # iterate through the list of strings and access string for each equation
    test = True 
    for equation in list_of_equations:
        # create a list with characters split up and identify parts of equation
        # get rid of spaces and join them back again into a single string with no spaces 
        splitstr = equation.split()
        characters = ""
        for item in splitstr:
            characters += item
        
        # for the different possible operators 
        if "+" in characters:
            characters1 = characters.split("+")
            n1 = int(characters1[0])
            characters2 = characters1[1].split("=")
            n2 = int(characters2[0])
            n3 = int(characters2[1])
            
            if n1 + n2 != n3: 
                test = False 
                
        elif "-" in characters: 
            characters1 = characters.split("-")
            n1 = int(characters1[0])
            characters2 = characters1[1].split("=")
            n2 = int(characters2[0])
            n3 = int(characters2[1])
            
            if n1 - n2 != n3:
                test = False
                
        elif "*" in characters: 
            characters1 = characters.split("*")
            n1 = int(characters1[0])
            characters2 = characters1[1].split("=")
            n2 = int(characters2[0])
            n3 = int(characters2[1])
            
            if n1 * n2 != n3:
                test = False
                
        elif "//" in characters:
            characters1 = characters.split("//")
            n1 = int(characters1[0])
            characters2 = characters1[1].split("=")
            n2 = int(characters2[0])
            n3 = int(characters2[1])
            
            if n1 // n2 != n3:
                test = False
        else: 
            characters1 = characters.split("%")
            n1 = int(characters1[0])
            characters2 = characters1[1].split("=")
            n2 = int(characters2[0])
            n3 = int(characters2[1])
            
            if n1 % n2 != n3:
                test = False
            
    return test 

print()
print('-' * 20)
print()

print('Test Case 1: check_math([" 3 +5 =9", "4  * 3 = 12 "])')

print()

result = check_math([" 3 +5 =9", "4  * 3 = 12 "])
print('Expected: False')
print('Actual:   ' + str(result))

print("Expected type of returned value: <class 'bool'>")
print('Actual type of returned value:   ' + str(type(result)))

print()
print('-' * 20)
print()

print('Test Case 2: check_math(["13 % 5 = 3", "4 // 3 = 1", "90 - 50 = 40"])')

print()

result = check_math(["13 % 5 = 3", "4 // 3 = 1", "90 - 50 = 40"])
print('Expected: True')
print('Actual:   ' + str(result))

print()
print('-' * 20)
print()

print('Test Case 3: check_math([])')

print()

result = check_math([])
print('Expected: True')
print('Actual:   ' + str(result))

print()
print('-' * 20)
print()
```
